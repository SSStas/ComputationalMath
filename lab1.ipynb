# Лабораторная работа №1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = 0.5
```

```python
h = np.array([2 / (2**n) for n in range(1, 21)])
```

```python
f1 = lambda x: np.sin(x**2)
df1 = lambda x: 2 * x * np.cos(x**2)

f2 = lambda x: np.cos(np.sin(x))
df2 = lambda x: -np.sin(np.sin(x)) * np.cos(x)

f3 = lambda x: np.exp(np.sin(np.cos(x)))
df3 = lambda x: -np.exp(np.sin(np.cos(x))) * np.sin(x) * np.cos(np.cos(x))

f4 = lambda x: np.log(x + 3)
df4 = lambda x: 1 / (x + 3)

f5 = lambda x: (x + 3)**0.5
df5 = lambda x: 0.5 / (x + 3)**0.5
```

```python
diff1 = lambda f, x, h: (f(x + h) - f(x)) / h
diff2 = lambda f, x, h: (f(x) - f(x - h)) / h
diff3 = lambda f, x, h: (f(x + h) - f(x - h)) / (2 * h)
diff4 = lambda f, x, h: (4 / 3) * (f(x + h) - f(x - h)) / (2 * h) - (1 / 3) * (f(x + 2 * h) - f(x - 2 * h)) / (4 * h)
diff5 = lambda f, x, h: 1.5 * (f(x + h) - f(x - h)) / (2 * h) - 0.6 * (f(x + 2 * h) - f(x - 2 * h)) / (4 * h) + 0.1 * (f(x + 3 * h) - f(x - 3 * h)) / (6 * h)
```

```python
funcs = [
    ('sin(x^2)', f1, df1), 
    ('cos(sin(x))', f2, df2), 
    ('exp(sin(cos(x)))', f3, df3), 
    ('ln(x + 3)', f4, df4), 
    ('(x + 3)^0.5', f5, df5),
]

diffs = [
    ('(f(x+h) - f(x))/h', diff1),
    ('(f(x) - f(x-h))/h', diff2),
    ('(f(x+h) - f(x-h))/2h', diff3),
    ('4/3 * (f(x+h) - f(x-h))/2h - 1/3 * (f(x+2h) - f(x-2h))/4h', diff4),
    ('3/2 * (f(x+h) - f(x-h))/2h - 3/5 * (f(x+2h) - f(x-2h))/4h + 1/10 * (f(x+3h) - f(x-3h))/6h', diff5),
]
```

```python
for funcName, f, df in funcs:
    plt.figure(figsize=(10, 6))
    plt.xlabel('h')
    plt.ylabel('Error value')
    plt.title(f'f(x) = {funcName} in x0 = {x0}')

    for diffName, diff in diffs:
        y = np.abs(diff(f, x0, h) - df(x0))
        plt.xscale('log')
        plt.yscale('log')
        plt.plot(h, y, label=diffName, marker='o')
    
    plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.4), fancybox=True)
    plt.grid(True)
    plt.show()
```
